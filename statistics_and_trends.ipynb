```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("data_set.csv")

# Adds company colummn to the dataframe from the car name
df["company"] = df.name.apply(lambda x : ' '.join(x.split(' ')[:1]))
```

```python
def create_relational_graph(df):
    """
    Creates a relational graph to show the relationship between 'selling_price' and 'km_driven'.

    Parameters:
    dataframe (pd.DataFrame): Input dataset.

    Returns:
    None
    """

    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='km_driven', y='selling_price', data=df, hue='company')
    plt.title('Selling Price vs Kilometers Driven')
    plt.xlabel('Kilometers Driven')
    plt.ylabel('Selling Price')
    plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()


create_relational_graph(df)
```

```python
def create_categorical_graph(df):
    """
    Creates a categorical graph to compare the distribution of 'fuel' categories.

    Parameters:
    dataframe (pd.DataFrame): Input dataset.

    Returns:
    None
    """
    plt.figure(figsize=(6, 6))
    sns.countplot(x='fuel', data=df,palette='Set2',hue='fuel')
    plt.title('Distribution of Fuel Types')
    plt.xlabel('Fuel Type')
    plt.ylabel('Count')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

create_categorical_graph(df)
```

```python
def create_statistical_graph(df):
    """
    Creates a statistical graph to show boxplots of 'selling_price' for each 'fuel' category.

    Parameters:
    dataframe (pd.DataFrame): Input dataset.

    Returns:
    None
    """

    plt.figure(figsize=(8, 6))
    sns.boxplot(x='fuel', y='selling_price', hue='fuel', data=df, palette='Set2', dodge=False)
    plt.title('Selling Price by Fuel Type')
    plt.xlabel('Fuel Type')
    plt.ylabel('Selling Price')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


create_statistical_graph(df)
```

```python
# Statistical analysis
numerical_columns = ['year', 'selling_price', 'km_driven']
print("\nStatistical Analysis:")
print(df[numerical_columns].describe())
print("\nCorrelation Matrix:")
print(df[numerical_columns].corr())
plt.figure(figsize=(8,6))
ax = sns.heatmap(df[numerical_columns].corr(), annot=True,annot_kws={'fontsize':15})
```
